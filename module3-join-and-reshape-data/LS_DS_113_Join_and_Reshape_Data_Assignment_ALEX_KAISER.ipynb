```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
#to compare results : table1
table1
```

```python
# table2 -> tidy
table2 = table2.reset_index()  #row index as new column
table2
```

```python
tidy = table2.melt(id_vars='index', value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])
tidy
```

```python
#rename columns
tidy = tidy.rename(columns={
    'index': 'trt', 
    'variable': 'name', 
    'value': 'result'
})
tidy
```

```python
#Replacing 'treatment in trt column
tidy.trt = tidy.trt.str.replace('treatment', '')

tidy
```

```python
# tidy -> table2
table2_new = tidy.pivot_table(index='name', columns='trt', values='result')
table2_new
```
